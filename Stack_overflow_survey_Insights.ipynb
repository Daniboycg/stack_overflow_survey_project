#### I just want to grab some specific columns from the original dataset. Instead of dropping tons of columns, I prefer to create a new df to work with. So, let's clean the data 

```python
interest_columns = ['ResponseId','MainBranch','Employment','RemoteWork','EdLevel','YearsCode','DevType','PurchaseInfluence','LanguageHaveWorkedWith','DatabaseHaveWorkedWith','PlatformHaveWorkedWith','OpSysProfessional use','OpSysPersonal use','Age','Gender','Ethnicity','WorkExp','ConvertedCompYearly']
len(interest_columns)
```

```python
df = original_df[interest_columns].copy()
df.head()
```

```python
df.set_index('ResponseId', inplace=True)
```

```python
df.rename(columns={
    'ConvertedCompYearly':'YearlySalary','DatabaseHaveWorkedWith':'WorkedDatabase',\
    'DevType':'CurrentJob','LanguageHaveWorkedWith':'WorkedLanguage',\
    'PlatformHaveWorkedWith':'WorkedPlatform','PurchaseInfluence':'Influence'},inplace=True)
```

```python
df.sort_index(axis='columns',inplace=True)
```

```python
df.head()
```
